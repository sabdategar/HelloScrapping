```python
# import BeautifulSoup first
from bs4 import BeautifulSoup
# import urlopen
from urllib.request import urlopen
```

```python
#this is the url that we want to scrapping
myUrl = 'https://www.tokopedia.com/'

from selenium import webdriver
```

```python
driver = webdriver.Chrome("/usr/lib/chromium-browser/chromedriver")
driver.get(myUrl)
```

```python
html = driver.page_source
print(html[0:1000])
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.title)
```

```python
product = soup.findAll('div',{'class':'scmnuRPZ'})
```

```python
from IPython.core.display import Image, display

pcName = '_2A1s174V'
psPB = 'zSWZBkL4'
pcPA = '_2B_PeQak'

for p in product:
    print('Nama Produk: ', p.find('div',{'class':pcName}).get_text())
    s = p.find('div',{'class':'_3WeQOLQl'})
    photo_url = s.find('img')['src']
    display(Image(photo_url, width=200, height=200,unconfined=True))
    print('Harga Asli: ', p.find('span',{'class':psPB}).get_text())
    print('Harga Promo: ', p.find('div',{'class':pcPA}).get_text())
    print(' ' * 500)
```
